# Models

```python
from sklearn.model_selection import GridSearchCV , cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix ,classification_report
```

* **train_test_split :**  Splits data into training and testing sets.
* **GridSearchCV :**  Finds best hyperparameters using cross-validation.
* **cross_val_score :**  Evaluates a model using cross-validation scores.

## Splitting the Data

```python
X_train = df.drop('Loan_Status',axis=1).values
y_train = df['Loan_Status'].values

X_test = df2.drop('Loan_Status',axis=1).values
y_test = df2['Loan_Status'].values
```

* x > input featurs
* y > target coloumn

## Scaling Data

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

for scaling features to range [0, 1]

```python
scaler2 = StandardScaler()
X_train = scaler2.fit_transform(X_train)
X_test = scaler2.transform(X_test)
```

it standardizes the features to have
* mean = 0 and standard deviation = 1.
* fits on the training data, then on the test data.

# ANN Model

## Creating the Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Dropout
```

* sequential : for adding sequential layers

* Dense : for creating Fully Connected Layer

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(units=14,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=7,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=3,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=1,activation='sigmoid'))

# For a binary classification problem
model.compile(loss='binary_crossentropy', optimizer='adam')
```

### two hidden layers

* 1st layer : has 30 neurons
* 2nd layer : has 15 neurons both use relu activation function

### Output layer :
 * for binary classification (two classes)

**Sigmoid activation:** Converts the neuron’s output into a probability between 0 and 1 thrishold = 0.5 which means output >= 0.5 is class 1 output < 0.5 is class 0

**binary_crossentropy:** the standard loss function for binary classification

**adam optimizer:** suitable for adaptive learning rate.

## Early Stopping

Stops training automatically when the metric stops improving

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', verbose=1, patience=25)
```

## Training the Model

```python
model.fit(x=X_train,
          y=y_train,
          epochs=300,
          validation_data=(X_test, y_test), verbose=1,
          callbacks=[early_stop])
```

* model.fit for training the model

* epochs = 300

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

model_loss.plot() : views a graph of loss updates occurs

## Testing the Model

```python
ANN_predicted = model.predict(X_test)
ANN_predicted = (ANN_predicted > 0.5).astype(int)
```

## Model Evaluation

```python
print("Accuracy:")
print(accuracy_score(y_test, ANN_predicted))
print("classification report:")
print(classification_report(y_test,ANN_predicted))
print("confusion_matrix:")
print(confusion_matrix(y_test,ANN_predicted))
```

* Calculates accuracy precision and recall .
* classification_report() :
Shows detailed performance metrics like precision, recall, and F1-score.
* confusion_matrix() :
Displays the precision and recall in a table format.

# Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
```

## Creating the Model

```python
Random_Forest = RandomForestClassifier(random_state=100)
RF_param_grid = {
    "n_estimators": [200,300, 400],
    "max_depth": [7, 10, 13],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2],
    "max_features": ["sqrt"],
    "class_weight": ["balanced"]
}
```

* RandomForestClassifier :
Creates a Random Forest model.
* param_grid = {} :
Defines a set of hyperparameters to be tested during tuning.
* "n_estimators":  Number of trees in the forest.
* "max_depth":  Maximum depth of each tree.
* "min_samples_split":  Minimum samples to split a node.
* "min_samples_leaf":  Minimum samples at a leaf node.
* "max_features":  Number of features considered when splitting nodes.
* "class_weight": ["balanced"]  Adjusts weights to handle imbalanced classes.

```python
RF_grid_search = GridSearchCV(
    estimator = Random_Forest,
    param_grid = RF_param_grid,
    cv=5,
    scoring="accuracy",
    verbose=1)
```

* grid_search = GridSearchCV() : >
 Creates an object of RF
* estimator, param_grid: >
Uses the Random Forest model and the defined hyperparameter grid.
* cv=5, scoring="accuracy" : >
Applies 5-fold cross-validation and evaluates using accuracy.

## Training the Model

```python
RF_grid_search.fit( X_train, y_train)
RF_best_model = RF_grid_search.best_estimator_
print(RF_grid_search.best_params_)
```

best_estimator_ :
Selects the model with the best hyperparameters after grid search.

## Testing the Model

```python
RF_best_model.fit(X_train, y_train)
RF_y_predicted = RF_best_model.predict(X_test)
```

* best_model.fit :
Trains the best model using the training data.

* best_model.predict(X_test) :
Uses the best trained model to predict results for the test data.

```python
RF_cv_scores = cross_val_score(
    RF_best_model,
    X_train,
    y_train,
    cv=5
)
```
